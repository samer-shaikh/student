```python
from pydantic import BaseModel
import joblib 
import pandas as pd

def build_features(data: pd.DataFrame) -> pd.DataFrame:
    data = data.copy()

    data.dropna(inplace=True)

    if "student_id" in data.columns:
        data.drop("student_id", axis=1, inplace=True)

    data = pd.get_dummies(data)

    return data
```

```python

class PredictionInput(BaseModel):
    student_id: int
    age: int
    gender: str
    course: str
    study_hours: float
    class_attendance: float
    internet_access: str
    sleep_hours: float
    sleep_quality: str
    study_method: str
    facility_rating: str
    exam_difficulty: str
```

```python
input_data = {
  "student_id": 4,
  "age": 22,
  "gender": "male",
  "course": "bsc",
  "study_hours": 7.4,
  "class_attendance": 85.3,
  "internet_access": "yes",
  "sleep_hours": 7.4,
  "sleep_quality": "average",
  "study_method": "coaching",
  "facility_rating": "medium",
  "exam_difficulty": "low"
}
```

```python
def predict_output(input_data: PredictionInput):

    df = pd.DataFrame([input_data])
    df_transformed = build_features(df)
    print(df_transformed)
    print('='*100)
    feature_columns = joblib.load("models/feature_columns.joblib")

    df_transformed = df_transformed.reindex(
        columns=feature_columns,
        fill_value=0
    )

    print(df_transformed)
    model = joblib.load('models/model.joblib')

    prediction = model.predict(df_transformed)

    result = int(prediction[0])

    return {'prediction':result}
```

```python
predict_output(input_data)
```
